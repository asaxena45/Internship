Q1. Write a python program to display all the header tags from‘en.wikipedia.org/wiki/Main_Page’.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("http://en.wikipedia.org/wiki/Main_Page")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
 print(soup.prettify())
```

```python
h1=soup.find_all('h1',class_="firstHeading")
h1
```

```python
h2=soup.find_all('h2',class_="mp-h2")
h2
```

```python
h3=soup.find_all('h3',class_="vector-menu-heading")
h3
```

```python
titles = soup.find_all(['h1', 'h2', 'h3'])
print('List all the header tags :', *titles, sep='\n\n')
```

```python
Titles=[]

for i in titles:
    Titles.append(i.text)
    
Titles
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({})
df['Titles']=Titles
df
```
