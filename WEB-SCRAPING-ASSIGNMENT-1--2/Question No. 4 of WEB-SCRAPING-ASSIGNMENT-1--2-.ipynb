```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://bookpage.com/reviews")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
book_name=soup.find_all('h4',class_="italic")
book_name
```

```python
Books_name=[]

for i in book_name:
    for j in i.find_all("a"):
        Books_name.append(j.text.replace("\n", " "))
        
Books_name
```

```python
author_name=soup.find_all('p', class_="sans bold")
author_name
```

```python
Authors_name=[]

for i in author_name:
    for j in i.find_all("p"):
        Authors_name.append(j.text.replace("</p>", " "))
        
Authors_name
```

```python
review=soup.find_all('p',class_="excerpt")
review
```

```python
Review=[]

for i in review:
    Review.append(i.text.replace("\n", " "))
        
Review
```

```python
genre=soup.find_all('p',class_="genre-links hidden-phone")
genre
```

```python
Genres=[]

for i in genre:
    for j in i.find_all("a"):
        Genres.append(j.text.replace("\n", " "))
        
Genres
```

```python
import pandas as pd
```

```python
Book=pd.DataFrame({})
Book['Book_name']=book_name[0:14]
Book['Author_name']=author_name[0:14]
Book['Review']=review[0:14]
Book['Genre']=genre[0:14]
Book
```
