Q10. Write a python program to scrape house details from https://www.nobroker.in/ for any location. It should 
include house title, location, area, emi and price.

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://www.nobroker.in/property/sale/delhi/Kailash%20Hills?searchParam=W3sibGF0IjoyOC41NTc3Mjg0LCJsb24iOjc3LjI1ODQ3NTUsInBsYWNlSWQiOiJDaElKOTdTZnNjRGpERGtScUpXOTYtZjBtYlkiLCJwbGFjZU5hbWUiOiJLYWlsYXNoIEhpbGxzIn1d&radius=2.0&propertyAge=0")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
print(soup.prettify())
```

```python
housetitles=soup.find_all("h2",class_="heading-6 font-semi-bold nb__1AShY")
housetitles
```

```python
House_titles=[]

for i in housetitles:
    House_titles.append(i.text.replace("\n", " "))
        
House_titles
```

```python
location=soup.find_all("div",class_="nb__2CMjv")
location
```

```python
Locations=[]

for i in location:
    Locations.append(i.text.replace("\n", " "))
        
Locations
```

```python
area=soup.find_all("div",class_="nb__3oNyC")
area
```

```python
Area=[]

for i in area:
    Area.append(i.text.replace("\n",""))

Area
```

```python
emi=soup.find_all("div",class_="font-semi-bold heading-6",id="roomType")
emi
```

```python
EMI=[]

for i in emi:
    EMI.append(i.text.replace("\n",""))

EMI
```

```python
price=soup.find_all("div",class_="font-semi-bold heading-6",)
price
```

```python
Price=[]
for i in price:
    for j in i.find_all("span"):
        Price.append(j.text.replace("\n",""))
Price[1:20:2]
```

```python
import pandas as pd
```

```python
House_details=pd.DataFrame({})
House_details["House_titles"]=House_titles
House_details["Locations"]=Locations
House_details["Area"]=Area
House_details["EMI"]=EMI
House_details["Price"]=Price[1:20:2]
House_details
```
